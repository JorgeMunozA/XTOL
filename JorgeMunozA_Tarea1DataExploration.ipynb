```python
import pandas as pd
import matplotlib as pt
%matplotlib inline
pt.style.use('ggplot')
data = pd.read_csv("defaultcreditcardclients.csv") 
```

```python
#Data Preparation
data['EDUCATION'].replace([1,2,3],['Graduate School','University','High School'],inplace=True)
data['EDUCATION'].replace([0,4,5,6],['Others','Others','Others','Others'],inplace=True)
data['MARRIAGE'].replace([1,2,3,0],['Married','Single','Divorce','Others'],inplace=True)
data['AGE Segments']=pd.cut(data['AGE'],bins=3,labels=["Young","Adult","Senior"])
data['Limit Segments']=pd.cut(data['LIMIT_BAL'],bins=3,labels=["Low","Medium","High"])
data['SEX'].replace([1,2],['male','female'],inplace=True)
data.rename(columns={'default payment next month': 'IsDefault'}, inplace=True)
```
